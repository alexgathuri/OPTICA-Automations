### Preqc

```python
bi.head(3)
```

```python
bi['dept'].unique()
```

```python
epreqc = bi[bi['dept']== 'PreQuality']
epreqc.head(3)
```

```python
epreqc['5 min'] = epreqc['Time Min'].apply(lambda x: 1 if x>5 else 0)
epreqc['15 min'] = epreqc['Time Min'].apply(lambda x: 1 if x>15 else 0)
epreqc.shape
```

```python
epreqc['start'] = pd.to_datetime(epreqc['start'], format = ('%Y/%m/%d %H:%M:%S'))
```

```python
epreqc['Hour'] = epreqc['start'].dt.strftime('%H')
epreqc.head(2)
```

```python
epreqcpivot1 = pd.pivot_table(epreqc, index='Hour', values = 'Doc Entry', aggfunc='count', fill_value = 0, margins= True, margins_name ='Grand Total')
epreqcpivot2 = pd.pivot_table(epreqc, index='Hour', values = 'Time Min', aggfunc='mean', fill_value = 0, margins= True, margins_name ='Grand Total')
epreqcpivot2 = epreqcpivot2.round(2)
epreqcpivot3 = pd.pivot_table(epreqc, index='Hour', values = '5 min', aggfunc='sum', fill_value = 0, margins= True, margins_name ='Grand Total')
```

```python
epreqcpivot1 = pd.merge(epreqcpivot1,epreqcpivot2, on='Hour', how= 'left' )
epreqcpivot1 = pd.merge(epreqcpivot1,epreqcpivot3, on='Hour', how= 'left' )
epreqcpivot1
```

```python
epreqcpivot1['% of Orders on Time'] =  (epreqcpivot1['Doc Entry'] - epreqcpivot1['5 min'])/epreqcpivot1['Doc Entry']
epreqcpivot1['% of Orders on Time'] = epreqcpivot1['% of Orders on Time'].map('{:.2%}'.format)
epreqcpivot1
```

```python
epreqcpivot1.rename(columns={'5 min':'Delays'}, inplace = True)
```

```python
epreqcpivot1=np.transpose(epreqcpivot1)
epreqcpivot1
```

```python
epreqcpivot11 = pd.pivot_table(epreqc, index='dept', values='5 min', aggfunc ='sum')
epreqcpivot22 = pd.pivot_table(epreqc, index='dept', values='15 min', aggfunc ='sum')
epreqcpivot11 = pd.merge(epreqcpivot11,epreqcpivot22, on='dept', how='left')
epreqcpivot11
```

```python
epreqc =epreqcpivot11
```
